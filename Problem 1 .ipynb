```julia
# Convert Texts to Padded Sequences
```

```julia
function text_to_indices(tokens::Vector{String}, word2idx::Dict{String, Int}, seq_len::Int)
    indices = [get(word2idx, word, word2idx["<UNK>"]) for word in tokens]

    if length(indices) > seq_len
        return indices[1:seq_len]
    end

    pad_len = seq_len - length(indices)
    return vcat(indices, fill(word2idx["<PAD>"], pad_len))
end
```

```julia
# BUILDING THE LSTM MODEL
```

```julia
using Flux
```

```julia
# Set constants
embed_size = 100  # size of word embedding vectors
```

```julia
hidden_size = 128  # number of hidden units in LSTM
```

```julia
output_size = 5  # number of classes "condition labels"
```

```julia
LSTM(embed_size => hidden_size)
```

```julia
Dense(hidden_size => output_size)
```

```julia
# Define model
LSTM_model = Chain(
    Embedding(vocab_size => embed_size),
    LSTM(embed_size => hidden_size),
    x -> dropdims(mean(x, dims=2), dims=2),
    Dense(hidden_size => output_size)
)
```

```julia
# Define the Loss Function and Optimizer
```

```julia
loss(xb, yb) = logitcrossentropy(LSTM_model(xb), onehotbatch(yb, 1:output_size))
```

```julia
opt = Adam(args.learning_rate)  
```

```julia
# Training Loop
```

```julia
using Flux
```

```julia
using Statistics: mean
```

```julia
using Flux: setup, update!
```

```julia
using Flux: onehotbatch, onecold, logitcrossentropy
```

```julia
opt_state = Flux.setup(Adam(args.learning_rate), LSTM_model)
```

```julia
for epoch in 1:args.epochs
    total_loss = 0.0
    batch_count = 0

    Flux.trainmode!(LSTM_model)

    for (xb, yb) in train_loader
        # Calculating the loss and gradients
        loss_val, grads = Flux.withgradient(LSTM_model) do m
            logitcrossentropy(m(xb), onehotbatch(yb, 1:output_size))
        end

        Flux.update!(opt_state, LSTM_model, grads[1])

        total_loss += loss_val
        batch_count += 1
    end

    avg_loss = total_loss / batch_count
    println("Epoch $epoch complete. Average loss: $avg_loss")
end
```

```julia
function calculate_confusion_matrix(preds, labels, n_classes)
    cm = zeros(Int, n_classes, n_classes)
    for (p, l) in zip(preds, labels)
        cm[p, l] += 1
    end
    return cm
end
```

```julia
function calculate_metrics(LSTM_model, loader)
    all_preds = Int[]
    all_labels = Int[]
    all_probs = Vector{Float32}[]
    
    Flux.testmode!(LSTM_model)
    for (xb, yb) in loader
        logits = LSTM_model(xb)
        probs = softmax(logits)
        preds = onecold(logits)
        
        append!(all_preds, preds)
        append!(all_labels, yb)
        append!(all_probs, eachcol(probs))
    end
    
    # Confusion matrix
    n_classes = maximum(all_labels)
    cm = calculate_confusion_matrix(all_preds, all_labels, n_classes)
    
    # Metrics
    accuracy = sum(diag(cm)) / sum(cm)
    
    # Per-class metrics
    class_metrics = []
    auc_scores = Float64[]
    
    for c in 1:n_classes
        tp = cm[c, c]
        fp = sum(cm[c, :]) - tp
        fn = sum(cm[:, c]) - tp
        tn = sum(cm) - tp - fp - fn
        
        precision = tp / (tp + fp + eps())
        recall = tp / (tp + fn + eps())
        f1 = 2 * precision * recall / (precision + recall + eps())
        
        push!(class_metrics, (
            precision = precision,
            recall = recall,
            sensitivity = recall,
            specificity = tn / (tn + fp + eps()),
            f1_score = f1
        ))
        
        # ===== Manual AUC Calculation (No ROCAnalysis.jl) =====
        y_true = all_labels .== c
        y_score = [p[c] for p in all_probs]
        
        # Sort by predicted score (descending)
        order = sortperm(y_score, rev=true)
        y_true_sorted = y_true[order]
        y_score_sorted = y_score[order]
        
        # Calculate TPR and FPR at each threshold
        tpr = cumsum(y_true_sorted) ./ sum(y_true_sorted)
        fpr = cumsum(.!y_true_sorted) ./ sum(.!y_true_sorted)
        
        # Add (0,0) and (1,1) to complete the curve
        tpr = [0; tpr; 1]
        fpr = [0; fpr; 1]
        
        # Compute AUC using trapezoidal rule
        auc_val = 0.0
        for i in 2:length(tpr)
            auc_val += (fpr[i] - fpr[i-1]) * (tpr[i] + tpr[i-1]) / 2
        end
        push!(auc_scores, auc_val)
        # ======================================================
    end
    
    return (
        accuracy = accuracy,
        class_metrics = class_metrics,
        confusion_matrix = cm,
        auc_scores = auc_scores
    )
end
```

```julia
function plot_roc_curves(metrics)
    
    plt = plot(title="ROC Curves by Class", 
               xlabel="False Positive Rate", 
               ylabel="True Positive Rate",
               legend=:bottomright)
    
    # Class names
    class_names = ["Neoplasms", "Digestive System Disease", "Nervous System Disease", "Cardiovascular Disease", "General Pathological Conditions"]
    
    # Plot each class
    for c in 1:5
        y_true = metrics.all_labels .== c
        y_score = [p[c] for p in metrics.all_probs]
        
        # Calculate ROC curve
        order = sortperm(y_score, rev=true)
        tpr = cumsum(y_true[order]) ./ sum(y_true)
        fpr = cumsum(.!y_true[order]) ./ sum(.!y_true)
        
        # Add (0,0) and (1,1)
        tpr = [0; tpr; 1]
        fpr = [0; fpr; 1]
        
        plot!(fpr, tpr, 
              label="$(class_names[c]) (AUC=$(round(metrics.auc_scores[c], 3)))",
              linewidth=2)
    end
    
    # Add reference line
    plot!([0, 1], [0, 1], linestyle=:dash, color=:black, label="Random Chance")
    return plt
end
```

```julia
function calculate_metrics_with_roc(LSTM_model, loader)
    all_preds = Int[]
    all_labels = Int[]
    all_probs = Vector{Float32}[]
    
    Flux.testmode!(LSTM_model)
    for (xb, yb) in loader
        logits = LSTM_model(xb)
        probs = softmax(logits)
        preds = onecold(logits)
        
        append!(all_preds, preds)
        append!(all_labels, yb)
        append!(all_probs, eachcol(probs))
    end
    
    n_classes = maximum(all_labels)
    roc_data = []
    
    for c in 1:n_classes
        y_true = all_labels .== c
        y_score = [p[c] for p in all_probs]
        
        # Sort by descending probability
        order = sortperm(y_score, rev=true)
        y_true_sorted = y_true[order]
        y_score_sorted = y_score[order]
        
        # Calculate TPR and FPR
        tpr = cumsum(y_true_sorted) ./ sum(y_true_sorted)
        fpr = cumsum(.!y_true_sorted) ./ sum(.!y_true_sorted)
        
        # Add (0,0) and (1,1)
        tpr = [0; tpr; 1]
        fpr = [0; fpr; 1]
        
        # Calculate AUC
        auc_val = sum((fpr[2:end] - fpr[1:end-1]) .* (tpr[2:end] + tpr[1:end-1])) / 2
        
        push!(roc_data, (fpr=fpr, tpr=tpr, auc=auc_val))
    end
    
    return roc_data
end
```

```julia
roc_data = calculate_metrics_with_roc(LSTM_model, test_loader)
```
